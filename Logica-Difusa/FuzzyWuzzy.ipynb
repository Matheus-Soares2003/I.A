```python
from fuzzywuzzy import fuzz
```

# **Similaridade de string em ordem**

Retornando a porcentagem de similaridade de strings

```python
fuzz.ratio('Apple Inc.', 'Apple')
```

```python
fuzz.partial_ratio("Apple Inc.", "Apple")
```

**Token_sort_ratio:** Ignora a ordem das palavras

```python
a = fuzz.token_sort_ratio("Lakers x Chicago Bulls", "Chicago Bulls x Lakers")
b = fuzz.ratio("Lakers x Chicago Bulls", "Chicago Bulls x Lakers")
c = fuzz.partial_ratio("Lakers x Chicago Bulls", "Chicago Bulls x Lakers")
print(a, b, c)
```
